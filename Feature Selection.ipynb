```python
import pandas as pd
import numpy as np
from sklearn.neighbors import NeighborhoodComponentsAnalysis
```

```python
prefix = 'not_denoised'
# prefix = 'denoised'
# prefix = 'noised'
# prefix = 'noise_denoised'
# Load your data into a pandas DataFrame
# Example assuming 'data' is your DataFrame with features and 'target' is your target variable
df = pd.read_csv(f'Feature Extracted/{prefix}_feature_extracted.csv')
df
```

```python
df_train = df[df['type'] == 'TRAIN'].copy().reset_index(drop=True)
df_val = df[df['type'] == 'VAL'].copy().reset_index(drop=True)
df_test = df[df['type'] == 'TEST'].copy().reset_index(drop=True)
```

```python
X_train = df_train.drop(columns=['Index', 'Final Label', 'Eye ID', 'type'])  # Features
y_train = df_train['Final Label']  # Target

X_val = df_val.drop(columns=['Index', 'Final Label', 'Eye ID', 'type'])  # Features
y_val = df_val['Final Label']  # Target

X_test = df_test.drop(columns=['Index', 'Final Label', 'Eye ID', 'type'])  # Features
y_test = df_test['Final Label']  # Target
y_val
```

```python
# Initialize NCA
nca = NeighborhoodComponentsAnalysis(n_components=10 ,random_state=42)
# nca = NeighborhoodComponentsAnalysis(random_state=42)

# Fit NCA to the data
nca.fit(X_train, y_train)
```

```python
# Optionally, you can retrieve the selected features
selected_features = np.argsort(nca.components_)

print("Selected features:", selected_features)
```

```python
# Transform the data using the learned transformation
X_train_transformed = nca.transform(X_train)

temp = pd.DataFrame(X_train_transformed)
temp['Final Label'] = y_train
temp

temp.to_csv(f'Feature Selected\{prefix}_feature_selected_train.csv', index=False)
```

```python
# Transform the data using the learned transformation
X_val_transformed = nca.transform(X_val)

temp = pd.DataFrame(X_val_transformed)
temp['Final Label'] = y_val
temp

temp.to_csv(f'Feature Selected\{prefix}_feature_selected_val.csv', index=False)
```

```python
# Transform the data using the learned transformation
X_test_transformed = nca.transform(X_test)

temp = pd.DataFrame(X_test_transformed)
temp['Final Label'] = y_test
temp

temp.to_csv(f'Feature Selected\{prefix}_feature_selected_test.csv', index=False)
```
